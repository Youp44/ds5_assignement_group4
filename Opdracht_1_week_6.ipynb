```python
import numpy as np 
import pandas as pd
import scipy.stats as sp
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
np.random.seed(2)

x = np.random.uniform(0,10,200)
y = 2 * x**2 - 5 * x + 3 + np.random.normal(0,10,200)

plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dataset")
plt.show()
```

```python
x_train = x[:80]
x_test = x[80:]

y_train = y[:80]
y_test = y[80:]

datatrain = {
    "x" : x_train,
    "y" : y_train
}

datatest = {
    "x" : x_test,
    "y" : y_test
}

datatrain = pd.DataFrame(datatrain)
datatest = pd.DataFrame(datatest)

X = sm.add_constant(datatrain["x"])
model = sm.OLS(datatrain["y"],X)
model = model.fit()

yhat = model.predict(X)

plt.scatter(datatrain["x"],datatrain["y"])
plt.plot(datatrain["x"], yhat, color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dataset")
plt.show()
```
